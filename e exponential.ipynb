## Compound interest

$$A = P \left(1+\frac{r}{n}\right) ^{nt}$$

and

$$\lim_{n \rightarrow \infty}P \left(1+\frac{r}{n}\right) ^{nt}=Pe^{rt}$$

## Radioactive decay

```python
Q_0,r,t = symbols('Q_0 r t',real=true)
Qt = Q_0*E**(r*t)
Qt
```

### Half life

```python
eq1=Eq(Qt.subs([(r,-0.0001238),(Q_0,1.0)]),0.5)
eq1
```

```python
solve(eq1,t)
```

### Population model

Current population growth rate is 1.3% per year.

Time to double population

```python
eq2=Eq(2.0,Qt.subs([(Q_0,1.0),(r,0.013)]))
eq2
```

```python
solve(eq2,t)[0].n(2)
```
