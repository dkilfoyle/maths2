```python
# setup
from sympy import *
x,y,z,p=symbols('x y z p')
init_printing()
%matplotlib notebook
```

## Price-Demand Equation

```python
pd = Eq(x,10000-1000*p)
pd
```

```python
#solve for price
solve(pd, p)[0]
```

```python
#revenue = price * units
revenue = _ * x
revenue
```

```python
cost = 7000 + 2*x
cost
```

## Profit

```python
profit = revenue - cost
profit.simplify()
```

```python
plot(revenue,cost, (x,0,10000))
```

```python
plot(profit,(x,0,10000))
```

## Maximise Profit

```python
diff(profit,x)
```
