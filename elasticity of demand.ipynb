```python
# setup
from sympy import *
x,y,z,p=symbols('x y z p',real=True)
init_printing()
%matplotlib notebook
```

## Demand decreases as price increases 

```python
eq1=Eq(x+500*p,10000)
eq1
```

```python
eq2=solve(eq1,x)[0]
Eq(x,eq2)
```

```python
plot(eq2,(p,0,20))
```

Revenue is price * demand

```python
price = solve(eq1,p)[0]
price
```

```python
revenue = price * x
revenue
```

```python
plot(revenue,(x,0,10000),xlabel="units",ylabel="revenue")
```
